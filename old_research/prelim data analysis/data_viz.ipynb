```plaintext
df = pd.read_csv("../All Earthquakes.csv")
df
```

```plaintext
# pull out var
size = df['Magnitude'] ** 3  # magnitude beign exponential for intensity 
lat = df['Longitude(E)']
long = df['Latitude(N)']

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    lat, 
    long, 
    s=size,  
    c=df['Depth'],           # depth color
    cmap='viridis',          # color palette
    alpha=0.2,               # transparency
    edgecolor='k'            # edge color for better visibility
)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Depth (km)')

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Visualization: Size (Magnitude) and Color (Depth)')

# Show the plot
plt.show()
```
